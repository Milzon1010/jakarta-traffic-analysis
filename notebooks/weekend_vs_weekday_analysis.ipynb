```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../data/jakarta_traffic_data.csv')

df['Date'] = pd.to_datetime(df['Date'])
df['Hari_Nama'] = df['Date'].dt.day_name()
df['Is_Weekend'] = df['Hari_Nama'].isin(['Saturday', 'Sunday'])
```

data weekend/weekday


```python
df['Hari_Nama'] = df['Date'].dt.day_name()
df['Is_Weekend'] = df['Hari_Nama'].isin(['Saturday', 'Sunday'])
```

## Total Kendaraan: Weekday vs Weekend
* Are vehicles more congested on weekdays?

```python
plt.figure(figsize=(6,4))
sns.barplot(x='Is_Weekend', y='Vehicle_Count', data=df, estimator='sum', ci=None)
plt.title('Total Kendaraan: Weekday vs Weekend')
plt.xlabel('Akhir Pekan?')
plt.ylabel('Jumlah Kendaraan')
plt.xticks([0, 1], ['Weekday', 'Weekend'])
plt.tight_layout()
plt.show()

#Answer: Are vehicles more congested on weekdays?
```

## Average Speed: Weekday vs Weekend
*   Are cars faster on weekends?

```python
plt.figure(figsize=(6,4))
sns.barplot(x='Is_Weekend', y='Average_Speed_kmh', data=df, estimator='mean', ci=None)
plt.title('Rata-rata Kecepatan: Weekday vs Weekend')
plt.xlabel('Akhir Pekan?')
plt.ylabel('Kecepatan (km/h)')
plt.xticks([0, 1], ['Weekday', 'Weekend'])
plt.tight_layout()
plt.show()
```

## Hourly Vehicle Number Pattern (divided by weekend/weekday)


```python
plt.figure(figsize=(10,6))
sns.lineplot(x='Hour', y='Vehicle_Count', hue='Is_Weekend', data=df, estimator='sum')
plt.title('Jumlah Kendaraan per Jam: Weekday vs Weekend')
plt.xlabel('Jam')
plt.ylabel('Jumlah Kendaraan')
plt.legend(labels=['Weekday', 'Weekend'])
plt.tight_layout()
plt.show()
```

#  Traffic Analysis: Weekday vs Weekend (Jakarta)

##  Objective

This analysis aims to compare traffic patterns in Jakarta between weekdays and weekends, focusing on:

- Total vehicle count
- Average vehicle speed
- Hourly traffic distribution

---

##  Dataset Overview

- **Total records:** 288 entries
- **Columns used:** `Date`, `Hour`, `Vehicle_Count`, `Average_Speed_kmh`, `Is_Weekend`, etc.
- **Data period:** Daily hourly traffic records in key Jakarta roads

---

##  1. Total Vehicle Count: Weekday vs Weekend

![Total Vehicle Count](../output/total_kendaraan_weekday_vs_weekend.png)

### Insight:
- The total number of vehicles on **weekdays is significantly higher** than on weekends.
- This reflects high mobility during working days (commute, deliveries, school).

---

##  2. Average Speed: Weekday vs Weekend

![Average Speed](../output/average_speed_weekday_vs_weekend.png)

###  Insight:
- Vehicles run **faster on weekends** due to less congestion.
- Average speed on weekdays drops to around **19 km/h**, while it reaches **30+ km/h** on weekends.

---

## 3. Hourly Traffic Pattern: Weekday vs Weekend

![Hourly Pattern](../output/hourly_vehicle_number_pattern_divided_by_weekend_weekday.png)

###  Insight:
- **Morning (79 AM)** and **evening (57 PM)** are clear rush hours on weekdays.
- On weekends, traffic volume is more evenly distributed throughout the day.
- Data gaps (e.g. hour 12) might indicate missing records and should be validated.

---

## Summary

| Metric                 | Weekday         | Weekend        |
|------------------------|-----------------|----------------|
| Total Vehicles         | High            | Low            |
| Average Speed (km/h)   | Lower (~19)     | Higher (~30)   |
| Traffic Peaks          | 79 AM, 57 PM  | Spread evenly  |
