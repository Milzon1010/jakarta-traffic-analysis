```python
df['Tahun'] = df['Date'].dt.year
df['Bulan'] = df['Date'].dt.month
df['Hari'] = df['Date'].dt.day

df['Hari_Nama'] = df['Date'].dt.day_name()

df[['Date', 'Hari_Nama', 'Bulan', 'Tahun']].head()
```

```python
df['Is_Weekend'] = df['Hari_Nama'].isin(['Saturday', 'Sunday'])
df['Jam_Sibuk'] = df['Hour'].apply(lambda x: 'Sibuk' if x in [7, 8, 9, 17, 18, 19] else 'Normal')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
sns.barplot(x='Hari_Nama', y='Vehicle_Count', data=df, estimator='sum', ci=None)
plt.title('Total Kendaraan per Hari')
plt.ylabel('Jumlah Kendaraan')
plt.xlabel('Hari')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x='Hour', y='Vehicle_Count', data=df, estimator='sum', ci=None)
plt.title('Total Kendaraan per Jam')
plt.ylabel('Jumlah Kendaraan')
plt.xlabel('Jam (0–23)')
plt.tight_layout()
plt.show()
```

```python
pivot = df.pivot_table(index='Hari_Nama', columns='Hour', values='Vehicle_Count', aggfunc='sum')
plt.figure(figsize=(12,6))
sns.heatmap(pivot, cmap='YlOrRd', linewidths=0.5, annot=True, fmt=".0f")
plt.title('Heatmap Jumlah Kendaraan per Hari dan Jam')
plt.ylabel('Hari')
plt.xlabel('Jam')
plt.tight_layout()
plt.show()
```
